```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('Power BI Updated Data analyst Data.xlsx')
```

## Q1. How many unique students are included in the dataset? 

```python
unique_students = df['Email ID'].nunique()
print('Unique Students:', unique_students)
```

### Conclusion: The dataset contains 2157 columns

## Q2. What is the average GPA of the students?

```python
average_gpa = df['CGPA'].mean()
print('Average GPA:', round(average_gpa, 2))
```

### Conclusion: The average CGPA of students is approximately 8.04.

## Q3. What is the distribution of students across different graduation years?

```python
graduation_distribution = df['Year of Graduation'].value_counts().sort_index()
print(graduation_distribution)

# Visualization
graduation_distribution.plot(kind='bar')
plt.title('Distribution of Students by Graduation Year')
plt.xlabel('Graduation Year')
plt.ylabel('Number of Students')
plt.show()
```

### Conclusion : Most students belong to the graduation years 2023 and 2024

## Q4. What is the distribution of student’s experience with Python programming?

```python
python_exp = df['Experience with python (Months)']
print(python_exp.describe())

# Visualization
sns.histplot(python_exp, bins=10, kde=True)
plt.title('Python Experience Distribution')
plt.xlabel('Months of Experience')
plt.ylabel('Count')
plt.show()
```

### Conclusion : Most students have 4–7 months of Python experience

## Q5. What is the average family income of the students?

```python
income_distribution = df['Family Income'].value_counts()
print(income_distribution)
```

### Conclusion: The majority of students belong to the 0–2 Lakh family income category.

## Q6. How does the GPA vary among different colleges? (Top 5 only)


```python
college_gpa = df.groupby('College Name')['CGPA'].mean().sort_values(ascending=False).head(5)
print(college_gpa)

# Visualization
college_gpa.plot(kind='barh')
plt.title('Top 5 Colleges by Average CGPA')
plt.xlabel('Average CGPA')
plt.ylabel('College')
plt.show()
```

### Conclusion : some colleges consistently show higher average CGPA values compared to others.

## Q7. Are there any outliers in the quantity attribute?

```python
# Boxplot for outlier detection
sns.boxplot(x=df['Quantity'])
plt.title('Outlier Detection in Quantity')
plt.show()
```

### Conclusion : The boxplot helps identify any abnormal values in the Quantity column.

## Q8. What is the average GPA for students from each city?

```python
city_gpa = df.groupby('City')['CGPA'].mean().sort_values(ascending=False)
print(city_gpa)

# Top 10 cities visualization
city_gpa.head(10).plot(kind='bar')
plt.title('Top Cities by Average CGPA')
plt.xlabel('City')
plt.ylabel('Average CGPA')
plt.show()
```

### Conclusion : Certain cities show higher average CGPA performance compared to others.

## Q9. Can we identify any relationship between family income and GPA?

```python
income_gpa = df.groupby('Family Income')['CGPA'].mean()
print(income_gpa)

# Visualization
income_gpa.plot(kind='bar')
plt.title('Family Income vs Average CGPA')
plt.xlabel('Family Income')
plt.ylabel('Average CGPA')
plt.show()
```

### Conclusion: The analysis helps determine whether family income has any influence on academic performance.

## Q10 How many students are from various cities?

```python
city_count = df['City'].value_counts().head(10)
print(city_count)

# Visualization
city_count.plot(kind='bar')
plt.title('Top 10 Cities by Student Count')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

### Conclusion : Some cities contribute significantly more students than others.

## Q11. How does the expected salary vary based on GPA, family income, and Python experience?

```python
# Scatter plot for GPA vs Salary
sns.scatterplot(x='CGPA', y='Expected salary (Lac)', data=df)
plt.title('CGPA vs Expected Salary')
plt.show()

# Scatter plot for Python Experience vs Salary
sns.scatterplot(x='Experience with python (Months)', y='Expected salary (Lac)', data=df)
plt.title('Python Experience vs Expected Salary')
plt.show()
```

## Conclusion: Students with higher CGPA and more Python experience generally expect higher salaries

## Q12. Which events attract more students from specific fields of study?

```python
event_designation = pd.crosstab(df['Events'], df['Designation'])
print(event_designation)

# Visualization
event_designation.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Events vs Field of Study')
plt.xlabel('Events')
plt.ylabel('Number of Students')
plt.show()
```

### Conclusion : Certain events are more popular among specific student groups.

## Q13. Do students with leadership skills tend to have higher GPA or expected salary?

```python
leadership_analysis = df.groupby('Leadership- skills')[['CGPA', 'Expected salary (Lac)']].mean()
print(leadership_analysis)
```

### Conclusion : Students with leadership skills often show better academic and salary expectations.

## Q14. How many students are graduating by the end of 2024?

```python
graduating_2024 = df[df['Year of Graduation'] <= 2024]['Email ID'].nunique()
print('Students graduating by 2024:', graduating_2024)
```

 ### Conclusion : This shows the number of students expected to graduate by the end of 2024 is 1664.

## Q15. Which promotion channel brings more student participation for the event?

```python
promotion_channel = df['How did you come to know about this event?'].value_counts()
print(promotion_channel)

# Visualization
promotion_channel.plot(kind='bar')
plt.title('Promotion Channel Effectiveness')
plt.xlabel('Promotion Channel')
plt.ylabel('Participation Count')
plt.xticks(rotation=45)
plt.show()
```

### The most effective promotion channel can be identified from the highest participation count is through Whatsapp

## Q16. Find the total number of students who attended Data Science related events.

```python
data_science_events = df[df['Events'].str.contains('Data Science', case=False, na=False)]

students_count = data_science_events['Email ID'].nunique()
print('Students attended Data Science events:', students_count)
```

### The result shows that 218 students were attended the event

## 17. Do students with high CGPA and more programming experience expect higher salaries?

```python
high_performance = df[(df['CGPA'] >= 8) & (df['Experience with python (Months)'] >= 6)]

average_salary = high_performance['Expected salary (Lac)'].mean()
print('Average Expected Salary:', round(average_salary, 2))
sns.scatterplot(
    x='Experience with python (Months)',
    y='Expected salary (Lac)',
    hue='CGPA',
    data=high_performance
)

plt.title('CGPA & Python Experience vs Expected Salary')
plt.show()
```

### Conclusion: Students with strong academic performance and more coding experience generally expect higher salary packages.
